```python
#Create function that accepts a row and creates a new column[employment status] based on occupation
census_df['Employment Status'] = ""

def employment_status(row):
    
    if 'Retired' in row['Occupation']:
        return 'Retired'
    elif 'Student' in row['Occupation']:
        return 'Student'
    elif 'Unemployed' in row['Occupation']:
        return 'Unemployed'
    elif row['Occupation'] == 'Child':  # Only consider exact match for "Child"
        return 'Child'
    elif row['Occupation'].lower().startswith('Child '):
        return 'Employed'
    else:
        return 'Employed'

#apply function to dataframe
    census_df['Employment Status'] = census_df.apply(employment_status, axis=1)
```

# Death rate

```python
#Calculate the death rates across teh age groups
age_groups = [(61,65), (56, 60), (66, 70), (71, 75), (76, 80), (81, 85), (86, 90), (91, 95), (96, 100), (101, 105), (106, 110),(111, 116)]

death_rates = []
for i in range(0, len(age_groups)-1):
    lower1, upper1 = age_groups[i]
    lower2, upper2 = age_groups[i+1]
    
    age_filter1 = (census_df['Age'] >= lower1) & (census_df['Age'] <= upper1) 
    age_filter2 = (census_df['Age'] >= lower2) & (census_df['Age'] <= upper2)
    
    pop1 = age_filter1.sum() 
    pop2 = age_filter2.sum()
    
    death_rate = (pop1 - pop2) / (upper2 - upper1)
    death_rates.append(death_rate)

print(death_rates)
```

```python
#high death rate compared to the average of 10.4 ((UK Death Rate 2021 | Statista, 2024))
population=len(census_df)
death_rate=(sum(death_rates)/population)*1000
death_rate
print(f'There are', round(death_rate), 'deaths per 1000 people')
```

# Population Growth
